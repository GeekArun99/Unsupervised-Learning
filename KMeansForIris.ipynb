```python
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import pandas as pd
```

```python
iris = load_iris()
```

```python
X = iris.data
y= iris.target
```

```python
#Visualize
sns.scatterplot(x=X[:,0], y=X[:,2])
```

```python
#Scaling the data 
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# optional - Dimensionality reduction using PCA

from sklearn.decomposition import PCA
pca =PCA(n_components=2)
pca_data = pca.fit_transform(X)
```

```python
#Elbow Method to find out the K values 
wcss =[]

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(pca_data)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
sns.lineplot(x=range(1,11), y=wcss, marker='o')
```

```python
#Kmeans
kmeans = KMeans(n_clusters=3, random_state =10)
labels = kmeans.fit_predict(pca_data)
```

```python
#plot with PCA data

sns.scatterplot(x=pca_data[:,0], y=pca_data[:,1], c=labels)
sns.scatterplot(x=kmeans.cluster_centers_[:,0], y=kmeans.cluster_centers_[:,1], marker="X", c="red", s=200)
```

```python
#plot witout PCA data

sns.scatterplot(x=pca_data[:,0], y=pca_data[:,1], c=y)
sns.scatterplot(x=kmeans.cluster_centers_[:,0], y=kmeans.cluster_centers_[:,1], marker="X", c="red", s=200)
```
